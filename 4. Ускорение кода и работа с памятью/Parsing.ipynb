```python
from bs4 import BeautifulSoup
html = get_html("https://coinmarketcap.com/")
soup = BeautifulSoup(html)
tds = soup.find('table', class_='h7vnx2-2 czTsgW cmc-table').find_all('a', class_='cmc-link')
links = []
for td in tds:
    try:
        a = td.get('href')
        link = 'https://coinmarketcap.com' + a
        links.append(link)
    except:
        continue
```
