# Создайте класс для первого интернет магазина. Функционал следующий:

## Со стороны владельца магазина:

- должен быть метод добавления товаров на склад (у каждого товара название, цена и количество)
- метод, который обновляет наличие товаров (на вход - ничего; на выход - ничего)
- метод, который считаем стоимость товаров по названию (на вход товар; на выход - стоимость)
- метод, который считаем стоимость товаров по категории (на вход - категория; на выход - стоимость)

## Со стороны пользователя сайта:
- метод, который показывает доступные товары по категории (на вход - категория; на выход - пара товар-цена)
- метод, который создает заказ (на вход - названия товаров и количество; на выход - стоимость корзины)

Названия товаров и категорий придумайте самостоятельно.
